# Introduction to Homography and Perspective Transformation
- Homography is a transformation ( a 3Ã—3 matrix ) that maps the points in one image to the corresponding points in the other image.
- Homograpy estimation from 4 points

```python
# Read image from file
import cv2
import numpy as np
import matplotlib.pyplot as plt 
img = cv2.imread('logo.png', cv2.IMREAD_UNCHANGED)
# resize image to 512x512 with respect to aspect ratio
img = cv2.resize(img, (512, 512), interpolation=cv2.INTER_AREA)
# Convert to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Check if image is loaded correctly
if img is None:
    print('Error: No image exists')
    exit(1)
# Convert to Grayscale
# img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display image
plt.imshow(img)
plt.show()
```

```python
# Keypoits for the image
kp = np.array([
    [120,400],
    [260,120],
    [380,400],
])

# Draw keypoints

for x,y in kp:
    cv2.circle(img, (x,y), 5, (255,0,0),thickness=-1)
    cv2.rectangle(img, (x-10,y-10), (x+10,y+10), (255,255,0),thickness=1)
# Define a function draw keypoints
def draw_keypoints(img, kp):
    """
    Draw keypoints on the image
    args:
        img: image
        kp: keypoints
    return:
        image with keypoints
    """
    for x,y in kp:
        cv2.circle(img, (x,y), 5, (255,0,0),thickness=-1)
        cv2.rectangle(img, (x-10,y-10), (x+10,y+10), (255,255,0),thickness=1)
    return img
# Display image
plt.imshow(img)
plt.show()
```
